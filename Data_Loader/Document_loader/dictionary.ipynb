```python
from langchain_community.document_loaders import DirectoryLoader,PyPDFLoader
```

```python
loader=DirectoryLoader(
    path='Book',
    glob='*.pdf',
    loader_cls=PyPDFLoader
)
docs=loader.load()
```

```python
docs=loader.load()
print(docs[300].page_content)
```

```python
# using Lazy_load object

loader=DirectoryLoader(
    path='Book',
    glob='*.pdf',
    loader_cls=PyPDFLoader
)
docs=loader.load()
# print(docs)
for document in docs:
    print(document.metadata)
    print("\n")
```

```python
# using Lazy_load object

loader1=DirectoryLoader(
    path='Book',
    glob='*.pdf',
    loader_cls=PyPDFLoader
)
docs=loader1.lazy_load()
# print(docs)
for document in docs:
    print(document.metadata)
    print("\n")
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
load_dotenv()

model = ChatGoogleGenerativeAI(
    model='gemini-2.5-flash'
)

prompt = PromptTemplate.from_template("What is the capital of {country}?")

chain = prompt | model | StrOutputParser()

chain.invoke({"country": "France"})
```

```python
promt_01=PromptTemplate(
    template = "Write a summary of that pdf file {pdf_file}",
    input_variables=["pdf_file"]
)

chain_01 = promt_01 | model | StrOutputParser()

result = chain_01.invoke({"pdf_file": "docs"})

print(result)

```
