```python
from langchain_community.utilities import WikipediaAPIWrapper

# Initialize the Wikipedia API wrapper
wikipedia_api_wrapper = WikipediaAPIWrapper(top_k_results=1, doc_content_chars_max=4000)
```

```python
# Create the WikipediaQueryRun tool
wikipedia_tool = WikipediaQueryRun(api_wrapper=wikipedia_api_wrapper)

# Invoke the tool to search for a topic
result = wikipedia_tool.invoke("LangChain")

# Print the retrieved information
print(result)
```
