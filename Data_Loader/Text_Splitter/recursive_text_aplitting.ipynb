```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text="""
This literature review presents a comprehensive analysis of potato leaf disease detection using modern deep learning techniques. 
Conventional machine-learning methods lack robustness in real-field conditions due to their dependence on handcrafted features. 
In contrast, Convolutional Neural Networks (CNNs) and transfer learning architectures demonstrate superior performance by 
learning discriminative visual patterns directly from image data. The study synthesizes existing research on datasets, augmentation 
strategies, architectures, field-based challenges, and PyTorch-based model implementations. It identifies key research gaps,
including the absence of real-world datasets, limited deployment-focused evaluations, and insufficient model interpretability. 
The review concludes by outlining recommendations for building robust, scalable, and deployable potato disease detection systems 
suitable for practical agricultural use.
"""
```

```python

splitter = RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=0,
    separators=["\n\n", "\n", " ", ""]
)

chunks = splitter.split_text(text)
```
