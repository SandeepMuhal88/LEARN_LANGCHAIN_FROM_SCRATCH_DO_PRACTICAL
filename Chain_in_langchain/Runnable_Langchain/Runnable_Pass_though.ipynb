```python
from langchain_core.runnables import RunnablePassthrough,RunnableSequence,RunnableParallel
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
load_dotenv()
```

```python
model = ChatGoogleGenerativeAI(
    model='gemini-2.5-flash-lite'
)
```

PromptTemplate

```python
prompt=PromptTemplate(
    template ='Generate Joke on that topic {topic}',
    input_variables=['topic']
)

prompt_1=PromptTemplate(
    template='Wire that joke explanation: {text}',
    input_variables=['text']
)
```

```python
parser=StrOutputParser()
```

```python
joke_generator=RunnableSequence(prompt,model,parser)
# text=joke_generator.invoke({"topic":"Cat"})

chain=RunnableParallel({
    'joke':RunnablePassthrough(),
    'explanation':RunnableSequence(prompt_1,model,parser)
})
```

```python
final_chain=RunnableSequence(joke_generator,chain)
final_chain.invoke({"topic":"Cat"})
```
