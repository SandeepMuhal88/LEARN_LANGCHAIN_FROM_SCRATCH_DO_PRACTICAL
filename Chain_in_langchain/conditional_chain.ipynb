```python
from langchain_huggingface import ChatHuggingFace,HuggingFaceEndpoint
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv

load_dotenv()

```

```python
# Load Model
llms=HuggingFaceEndpoint(
    repo_id="google/gemma-2-2b-it",
    task="text-generation"
)
model1=ChatHuggingFace(llm=llms)
model2=ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite",
    temperature=0
)
```

```python
parser=StrOutputParser()
```

```python
prompt=PromptTemplate(
    template="classify the sentiment of the following feedback text into positive and Negative \n {feedback}",
    input_variables=["feedback"]
)

chain=prompt | model1 | parser

result=chain.invoke({"feedback":"The product was great! I would buy it again."})
print(result)
    
```
