```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter 
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_overlap=0,
    separators=["\n\n", "\n", " ", ""],
)
```

```python
# 1. Initialize the Loader (Only pass the file path here)
# Note: Use raw string r"..." or double backslashes \\ for Windows paths
loader = PyPDFLoader(r"D:\Project-to-learn\Learn_Langchain\Vector_Store\Cleaned Ml Roadmap.pdf")

# 2. Load the raw documents
raw_docs = loader.load()
```

```python
# 3. Split the documents
final_docs = splitter.split_documents(raw_docs)

print(f"Loaded {len(raw_docs)} pages.")
print(f"Split into {len(final_docs)} chunks.")
```

```python
from langchain_core.embeddings import Embeddings
```

```python
embeddinggs= Embeddings
```

```python
from langchain_community.vectorstores import FAISS

vectorstore = FAISS.from_documents(raw_docs, embeddinggs)
```
