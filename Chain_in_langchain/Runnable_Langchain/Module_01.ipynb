Solution useing Abstraction method

```python
from abc import ABC, abstractmethod
```

```python
class Runnable(ABC):
    @abstractmethod
    def invoke(input_data):
        pass
```

```python
import random
class NakliLLM(Runnable):
    def __init__(self):
        print("LLMS Created")



    def invoke(self,prompt):
        responces=[
            'Delhi is capital of india',
            'Mumbai is capital of maharashtra',
            'Bangalore is capital of Karnataka',
            'Hyderabad is capital of Telangana',
            'Chennai is capital of Tamil Nadu'
        ]    
        return {'response':random.choice(responces)}

    def predict(self,prompt):

        responces=[
            'Delhi is capital of india',
            'Mumbai is capital of maharashtra',
            'Bangalore is capital of Karnataka',
            'Hyderabad is capital of Telangana',
            'Chennai is capital of Tamil Nadu'
        ]    
        return {'response':random.choice(responces)}

    

class Nakli_PromptTemplate(Runnable):
    def __init__(self,template,input_variables):
        self.template=template
        self.input_variables=input_variables

    def invoke(self,input_dict):
        return self.template.format(**input_dict)   
    def format(self,input_dict):
        return self.template.format(**input_dict)
        
       

class NakaliLLMsChain():
    def __init__(self,llms,prompt):
        self.llms=llms
        self.prompt=prompt
    
    def runable(self,input_dict):
        final_prompt=self.prompt.format(input_dict)
        result=self.llms.predict(final_prompt)
        return result['response']


class Nakali_Stroutput_parsers(Runnable):
    def __init__(self):
        pass

    def invoke(self,input_dict):
        return input_dict
```

```python
class Runnable_cannactor(Runnable):
    def __init__(self, runnable_list):
        self.runnable_list = runnable_list
    
    def invoke(self,input_data):
        for runnable in self.runnable_list:
           input_data = runnable.invoke(input_data)      

        return input_data
```

```python
template= Nakli_PromptTemplate(
    template="what is the capital of {country} and explain that {Topic}",
    input_variables=["country","Topic"]
)
```

```python
parser=Nakali_Stroutput_parsers()
```

```python
llm= NakliLLM()
```

```python
chain=Runnable_cannactor([template,llm,parser])
```

```python
chain.invoke({'country': 'India','Topic':'History'})
```
