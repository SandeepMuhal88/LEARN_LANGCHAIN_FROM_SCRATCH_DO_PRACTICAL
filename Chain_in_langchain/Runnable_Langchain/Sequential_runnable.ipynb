```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.runnables import RunnableSequence
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
model=ChatGoogleGenerativeAI(model='gemini-2.0-flash')
```

```python
prompt_template=PromptTemplate(
    template='make a report about {Topic}',
    input_variables=['Topic']
)

prompt_template_01=PromptTemplate(
    template='make a 5 point of summary of {text}',
    input_variables=['text']
)
```

```python
parser=StrOutputParser()

chain=RunnableSequence(prompt_template,model,parser,prompt_template_01,model,parser)
```

```python
# print(chain.invoke({'Topic':'black hole'}))
responce=chain.invoke({'Topic':'black hole'})
print(responce)
```

```python
chain.get_graph().print_ascii()
```
