```python
from langchain_core.prompts import PromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_huggingface import HuggingFaceEndpoint,ChatHuggingFace
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
load_dotenv()
```

```python
from transformers.models import m2m_100


llms=HuggingFaceEndpoint(
    repo_id="google/gemma-2-2b-it",
    task="text-generation"
)
m1=ChatHuggingFace(llm=llms)

m2=ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite",
    temperature=0
)
parser=StrOutputParser()
```

```python
from langchain_core import prompts


prompt_01=PromptTemplate(
    template="Generate  short and simple notes from the following topic: \n {topic}",
    input_variables=["topic"],
)
prompt_02=PromptTemplate(
    template="Generate  5 short and simple question and answer from the  following topic: \n {topic}",
    input_variables=["topic"],
)

prompt_03=PromptTemplate(
    template="Merge the provided notes and quiz into a single document \n notes: {notes} \n quiz: {quiz}",
    input_variables=["notes","quiz"],
)
```

```python
parsers=StrOutputParser()
```

```python
from langchain_core.runnables import RunnableParallel
```

```python
# Define Chain method

parallel_chain = RunnableParallel({
    "notes": prompt_01 | m1 | parser,
    "quiz": prompt_02 | m2 | parser,
})

merge_chain = prompt_03 | m2 | parser

# Full pipeline
final_chain = parallel_chain | merge_chain

print(final_chain.invoke({"topic": "Transformers"}))
```

```python
final_chain.get_graph().print_ascii()
```
