Make Parallel Chain in langchain

```python
# from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_huggingface import HuggingFaceEndpoint,ChatHuggingFace
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
load_dotenv()
```

```python
from transformers.models import m2m_100


llms=HuggingFaceEndpoint(
    repo_id="google/gemma-2-2b-it",
    task="text-generation"
)
m1=ChatHuggingFace(llm=llms)

m2=ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite",
    temperature=0
)
parser=StrOutputParser()
```

```python
temp_01=PromptTemplate(
    template="Explain me summary of that {topic}",
    input_variables=["topic"]
)
temp_02=PromptTemplate(
    template="Explain me this hidden knowledge {text}",
    input_variables=["text"]
)
temp_03=PromptTemplate(
    template="Base on text that is infomation is very systhesis {text_2}",
    input_variables=["text_2"]

)
```

```python
chain_1=temp_01|m1|parser
text=chain_1.invoke({'topic':'AI'})
print(text)
```

```python
chain_2=temp_02|m2|parser
text_2=chain_2.invoke({'text':'text'})
print(text_2)
```

```python
chain=temp_03|m2|parser
result=chain.invoke({'text_2':'text_2'})
print(result)
```
