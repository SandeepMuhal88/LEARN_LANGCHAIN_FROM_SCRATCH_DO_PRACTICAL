With Hugging Face open source model

```python
from langchain_huggingface import ChatHuggingFace,HuggingFaceEndpoint
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
from dotenv import load_dotenv
load_dotenv()
```

```python
llm=HuggingFaceEndpoint(
    repo_id="google/gemma-2-2b-it",
    task="text-generation"
)
```

```python
model=ChatHuggingFace(
    llm=llm
)
```

Make Template 

```python
prompt_01=PromptTemplate(
    template='Write a 5 line summary on the following text. /n {text}',
    input_variables=['text']
)

prompt_02=PromptTemplate(
    template='Write a 5 points summary on the following text. /n {text}',
    input_variables=['text']
)

parsers=StrOutputParser()
```

```python
chain=prompt_01 | model | parsers | prompt_02 | model | parsers
response=chain.invoke({'text':'black hole'})
print(response)
```

```python
chain.get_graph().print_ascii()
```
