```python
from langchain_core.tools import tool
```

USING @ tool decorator

```python
#Step-01 Create a FUnction 

def multiply(a,b):
    """Multuply two numbers """
    return a*b
```

```python
# Step-02 Add type hints

def multiply(a:int,b:int)->int:
    """Multiply two numbers"""
    return a*b
```

```python
# Step-03 setup tool decorator

@tool
def multiply(a:int,b:int)->int:
    """Multiply two numbers"""
    return a*b

```

```python
result=multiply.invoke({"a":5,"b":7})
print(result)
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

```python
print(multiply.args_schema.model_json_schema())
```

Using StructuredTool

```python
from pydantic import BaseModel,Field
from langchain_core.tools import StructuredTool
```

```python
class MultiplyInput(BaseModel):
    a: int = Field(required=True, description="The first number to add")
    b: int = Field(required=True, description="The second number to add")
```

```python
def multiply_func(a: int, b: int) -> int:
    return a * b
```

```python
multiply_tool = StructuredTool.from_function(
    func=multiply_func,
    name="multiply",
    description="Multiply two numbers",
    args_schema=MultiplyInput
)
```

```python
result = multiply_tool.invoke({'a':3, 'b':3})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

USING BASETOOL Class

```python
from langchain_core.tools import  BaseTool
from typing import Type
```

```python
#ARG Schema using pydantic
```

```python
class MuliplyINPUT(BaseModel):
    a:int=Field(required=True,description="The First Number is add")
    b:int=Field(required=True,description="The Second Number is add")
```

```python
class MultiplyTools(BaseTool):
    name:str="Multiply"
    description:str="Multiply two numbers"
    arg_schema:Type[BaseModel]=MuliplyINPUT
    def _run(self,a:int,b:int)->int:
        return a*b
```

```python
multiply_tool=MultiplyTools()
```

```python
result = multiply_tool.invoke({'a':3, 'b':3})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)

print(multiply_tool.args)
```
